# Examining Raw Waveform Files (.m files)

Welcome! In our previous notebooks, we looked at a catalog (a list of earthquake events and times). Now, we are going to look at the **actual recordings** of the earthquakes.

These recordings are called **waveforms**. They represent the physical vibrations (wiggles) that the seismometer felt in the ground. 

In our folder, these waveforms are saved as `.m` files. Each file contains a snippet of recording for a specific earthquake event. Let's open one up and see what's inside!

### Step 1: Importing Our Tools
We need a few tools:
- `obspy`: Our trusty seismology toolkit. It can read these waveform files effortlessly.
- `glob` and `os`: Standard Python tools for finding files inside folders on your computer.

```python
import obspy
import glob
import os
```

### Step 2: Locating the Files
First, we need to tell Python where our files are saved. We use `glob` to search for anything that ends in `.m` inside our folder.

```python
# Define the path to the folder holding the files
waveform_dir = 'top_300_raw_cut_waveforms/'

# Create a search pattern for files ending in .m
file_pattern = os.path.join(waveform_dir, '*.m')

# Find all matching files and store them in a list
m_files = glob.glob(file_pattern)

print(f"Searching in: {waveform_dir}")
print(f"Successfully found {len(m_files)} waveform files!")
```

### Step 3: Reading a Waveform File
Let's just take the very first file we found and read it using `obspy`.

When `obspy` reads a waveform file, it returns a **Stream**. You can think of a Stream as a bundle of wires. Each "wire" is called a **Trace**, and each Trace represents the continuous recording from one single sensor component (like up/down or North/South movement).

```python
if m_files:
    # Grab the first file in our list
    first_file = m_files[0]
    print(f"Opening file: {os.path.basename(first_file)}")
    
    # Tell obspy to read the file!
    st = obspy.read(first_file)
    
    # How many "wires" (Traces) are in this bundle (Stream)?
    print(f"Total Traces inside this file: {len(st)}")
else:
    print("Oops, no files were found!")
```

### Step 4: Inspecting the Metadata (The `stats` dictionary)
Every Trace has two main parts:
1. **`.data`**: A giant list of numbers representing the actual physical wiggles.
2. **`.stats`**: A handy dictionary attached to the Trace that acts like a sticky note, telling us all the important background information (metadata) about the recording.

Let's loop through just the first 3 Traces in our file and print out what's written on their `stats` "sticky notes"!

```python
if m_files:
    # Look at the first 3 traces only
    for j, tr in enumerate(st[:3], start=1):
        print(f"--- Trace {j} Info ---")
        
        # Where did this recording come from?
        print(f"Station Code:  {tr.stats.station}")
        print(f"Network Code:  {tr.stats.network}")
        
        # What direction was the sensor facing? (e.g., HHZ is usually vertical)
        print(f"Channel:       {tr.stats.channel}")
        
        # When did the recording start and end?
        print(f"Start Time:    {tr.stats.starttime}")
        print(f"End Time:      {tr.stats.endtime}")
        
        # How fast was the sensor recording? (e.g., 200 Hz means 200 data points per second!)
        print(f"Sampling Rate: {tr.stats.sampling_rate} Hz")
        print(f"Data Points:   {tr.stats.npts}")
        
        # Let's do some math to find out how long the recording is in total seconds
        total_seconds = tr.stats.endtime - tr.stats.starttime
        print(f"Total Length:  {total_seconds:.2f} seconds")
        print("")
```

---
**Summary:**
As you can see, a single `.m` file is actually a rich mini-database containing pre-cut, synchronized windows of waveform data. It doesn't just hold the data; it holds everything we need to know *about* the data, all thanks to the `stats` dictionary attached to every Trace!
