# Examining QuakeML Data: A First Look

Welcome! In this notebook, we are going to learn how to open a file containing raw earthquake data (called a **QuakeML** file) and just peek inside to see what it looks like. 

Before we do any complex data cleaning or organizing, it's always a good idea to just open the "box" and examine the raw information. We will look at where the earthquake happened, how big it was, and the raw timestamps (called **picks**) of when the seismic waves hit our recording stations.

### Step 1: Importing Our Tool
We only need one tool for this task:
- `obspy`: A powerful Python library built specifically for seismology. It knows how to translate complex earthquake file formats into Python objects we can easily read.

```python
import obspy
```

### Step 2: Loading the Earthquake Catalog
We tell `obspy` where our XML file is, and it reads all the earthquake events into a big list called a **Catalog**.

```python
# Define the path to our data file
xml_file = "filtered_events.xml"

# Read the file into a catalog
cat = obspy.read_events(xml_file)

print(f"Number of earthquakes found in this file: {len(cat)}")
```

### Step 3: Examining a Single Earthquake
Let's grab just the very first earthquake in our catalog so we don't overwhelm our screen with text!

We will extract its core properties:
- **Event ID**: The unique name tag for this earthquake.
- **Origin Time**: The exact moment the earth ruptured.
- **Location**: The Latitude and Longitude coordinates.
- **Depth**: How deep the rupture was.
- **Magnitude**: How much energy the earthquake released (its "size").

```python
# Grab the first event to examine
event = cat[0]

print(f"--- Examining Event ---")
print(f"Event ID: {event.resource_id}")

# Check if the event has origin data (location/time)
if event.origins:
    print(f"Origin Time: {event.origins[0].time}")
    print(f"Latitude: {event.origins[0].latitude}")
    print(f"Longitude: {event.origins[0].longitude}")
    print(f"Depth: {event.origins[0].depth} meters")

# Check if the event has a calculated magnitude
if event.magnitudes:
    print(f"Magnitude: {event.magnitudes[0].mag}")
else:
    print("Magnitude: Not available")
```

### Step 4: Looking at the Raw Wave Picks
When an earthquake happens, vibrations travel outwards and are recorded by scientific stations.
The exact time a wave arrives at a station is called a **pick**.

Because computer algorithms and human scientists might both try to label the waves, a single station might have multiple picks for the exact same wave! Here, we are just printing out *everything* raw, without any filtering, so you can see how messy raw data can sometimes be.

```python
if event.picks:
    print("Raw Wave Arrivals (Picks):")
    for pick in event.picks:
        # Get the station name (if it exists)
        station = pick.waveform_id.station_code if pick.waveform_id else "Unknown"
        
        # Get the method used to make this pick (e.g., 'modelled' by a human, or 'autopick' by a computer)
        method = str(pick.method_id) if pick.method_id else "Unknown"
        
        print(f"  - Station: {station}, Phase (Wave Type): {pick.phase_hint}, Time: {pick.time}, Method ID: {method}")
else:
    print("No picks found for this event.")
```
